## 1) Load the data set into Pandas, and define a data frame just including columns ‘time_in_hospital’, ‘insulin’ and ‘num_lab_procedures’. 

```python
# Importing pandas and numpy libraries
import pandas as pd
import numpy as np
```

```python
# Load dataset and change all '?' to null values
missing_value=["?",np.nan]
df=pd.read_csv('diabetic_data.csv',na_values=missing_value,low_memory=False)
```

```python
# View dataframe top rows
df.head()
```

```python
# Create dataframe df2 and get all rows and columns 'time_in_hospital','insulin' and 'num_lab_procedures'
df2= df.loc[:,["time_in_hospital","insulin","num_lab_procedures"]]
df2
```

## 2) Use a method to remove the missing data in ‘time_in_hospital’ column, and perform z-score normalization of the values in this column and generate a new column that contains the normalised values.

```python
# fill all null values to 0
df2['time_in_hospital'].fillna(0,inplace=True)
```

```python
# Shows whether is there any null values in column 'time_in_hospital'
df2['time_in_hospital'].isnull().sum()
```

```python
# Get time_in_hospital column
zscorehospital=df.loc[:,"time_in_hospital"]
```

#### Z-Score is computed by subtracting each value by the mean of the corresponding column and then dividing by the standard deviation1

```python
# calculate z-score for time_in_hospital and insert in column 2
df2.insert(1,"z-scorehospital",(zscorehospital-np.mean(zscorehospital))/np.std(zscorehospital),True)
df2
```

## 3)Perform a one-hot-encoding of the attribute values in column ‘insulin’. To do so, first identify the list of unique values in this column (considering missing values as a special category), and then replace each of those values by a unique one-hot encoding vector (e.g.., replace all occurrences of ‘No’ by [1, 0, 0, 0, 0, 0]). Generate new column that contains the one-hot encoding vectors.

```python
# fill all null values to 0
df2['insulin'].fillna('special_category',inplace=True)
```

```python
#perform ohe for column insulin
ohe= pd.get_dummies(df2['insulin'],prefix='Category')
print(ohe)
ohe['insulin_ohe'] = ohe[['Category_Down','Category_No','Category_Steady','Category_Up',]].values.tolist()
```

#### For insulin values Down is [1,0,0,0] , No is [0,1,0,0] , Steady is [0,0,1,0] , Up is [0,0,0,1]

```python
# Insert the one hot encoding into column 4
df2.insert(3,"insulin_ohe",ohe["insulin_ohe"],True)
df2
```

### 4) Create 5 bins for the column ‘num_lab_procedures’, and generate a new column to contain the bin values. Also explain how you determine the size of bins

```python
# Find out the max number and min num for num_lab_procedures
num_lab_procedure_col=df.loc[:,"num_lab_procedures"]
print(num_lab_procedure_col.max())
print(num_lab_procedure_col.min())
```

#### The range of column 'num_lab_procedures is 1-132, to create 5 bins for this column i did 132/5 which is 26.4 and i rounded it up to 26.
#### From there i create different labels for each range, 1-26 belong to 1, 26-52 belong to 2, 53-78 belongs to 3, 79-104 belongs to 4 and 105 above belong to 5

```python
#Create range for each bins and the label and insert the data as column num_lab_porocedures_category
bins=[0,26,52,78,104,132] 
label=[1,2,3,4,5]
df2['num_lab_procedures_category']=pd.cut(df2['num_lab_procedures'],bins,labels=label)
df2
```
